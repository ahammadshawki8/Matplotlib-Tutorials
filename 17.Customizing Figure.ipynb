```python
# module 17

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
fig1, ax1=plt.subplots(ncols=2,nrows=2,constrained_layout=True)
# constrained_layout ensures some spacing.
# we can also use tight_layout() command instead of that.
```

```python
fig2=plt.figure(constrained_layout=True)
# it will create an empty figure
spec2=gridspec.GridSpec(ncols=2,nrows=2, figure=fig2)
# it will make the figure for comfortable with 2 rows and 2 columns.

# manually adding subplot with spec2 position.
ax2=fig2.add_subplot(spec2[0,0])
ax3=fig2.add_subplot(spec2[0,1])
ax4=fig2.add_subplot(spec2[1,0])
ax5=fig2.add_subplot(spec2[1,1])

# here the first position is [column,row]
# column starts counting from up to down.
# row starts counting from left to right.

# Note:
# While indexing in NumPy follows the form (row, column),
# the matplotlib follows the form (width, height)/(column,row).
```

```python
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(3, 3)
# we can also use add_gridspec to create gridspec.

f3_ax1 = fig3.add_subplot(gs[0, :])
f3_ax1.set_title('gs[0, :]')

f3_ax2 = fig3.add_subplot(gs[1, :2])
f3_ax2.set_title('gs[1, :2]')

f3_ax3 = fig3.add_subplot(gs[1:, 2])
f3_ax3.set_title('gs[1:, 2]')

# we can also use the negative indexing.
f3_ax4 = fig3.add_subplot(gs[-1, 0])
f3_ax4.set_title('gs[-1, 0]')

f3_ax5 = fig3.add_subplot(gs[-1, -2])
f3_ax5.set_title('gs[-1, -2]')
```

```python
fig5=plt.figure(constrained_layout=True)
spec4=fig5.add_gridspec(nrows=2,ncols=2)

annotate_options=dict(xy=(0.5,0.5), xycoords="axes fraction", va="center", ha="center")

ax1=fig5.add_subplot(spec4[0,0])
ax1.annotate("GridSpec[0,0]", **annotate_options)

ax2=fig5.add_subplot(spec2[1,0]).annotate("GridSpec[1,0]", **annotate_options)
ax3=fig5.add_subplot(spec2[1,1]).annotate("GridSpec[1,1]", **annotate_options)
ax4=fig5.add_subplot(spec2[0,1]).annotate("GridSpec[0,1]", **annotate_options)
```

```python
# just for fun making a complex plot.

import numpy as np
from itertools import product


def squiggle_xy(a, b, c, d, i=np.arange(0.0, 2*np.pi, 0.05)):
    return np.sin(i*a)*np.cos(i*b), np.sin(i*c)*np.cos(i*d)


fig11 = plt.figure(figsize=(8, 8), constrained_layout=False)

# gridspec inside gridspec
outer_grid = fig11.add_gridspec(4, 4, wspace=0.0, hspace=0.0)# creating 16 axes with no spaces

for i in range(16):
    inner_grid = outer_grid[i].subgridspec(3, 3, wspace=0.0, hspace=0.0)
    a, b = int(i/4)+1, i % 4+1
    for j, (c, d) in enumerate(product(range(1, 4), repeat=2)):
        ax = fig11.add_subplot(inner_grid[j])
        ax.plot(*squiggle_xy(a, b, c, d))
        ax.set_xticks([])# making the ticks invisible. ticks mean the numbering of an axis.
        ax.set_yticks([])
        
all_axes = fig11.get_axes()# grabbing all axes situated in the figure
product(range(1, 4), repeat=2)

# show only the outside spines
# setting up the borders.
for ax in all_axes:
    for sp in ax.spines.values():
        sp.set_visible(False)
    if ax.is_first_row():
        ax.spines['top'].set_visible(True)
    if ax.is_last_row():
        ax.spines['bottom'].set_visible(True)
    if ax.is_first_col():
        ax.spines['left'].set_visible(True)
    if ax.is_last_col():
        ax.spines['right'].set_visible(True)

plt.show()
```
