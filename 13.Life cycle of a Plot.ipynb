```python
# module 13
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from matplotlib.ticker import FuncFormatter
%matplotlib inline
import pandas as pd
```

```python
data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}

group_names=np.array(list(data.keys()))
group_data=np.array(list(data.values()))
group_mean=np.mean(group_data)
print(group_mean)
```

```python
plt.style.use("fivethirtyeight")
fig,ax=plt.subplots()
ax.barh(group_names,group_data,color="grey")
```

```python
plt.rcParams.update({"figure.autolayout":True})

fig,ax=plt.subplots()
ax.barh(group_names,group_data)

labels=ax.get_xticklabels()
plt.setp(labels,rotation=45,horizontalalignment="right")

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')

plt.show()
```

```python
# We can also adjust the size of this plot using the pyplot.subplots() function. We can do this with the figsize kwarg.
# Note:
# While indexing in NumPy follows the form (row, column),
# the figsize kwarg follows the form (width, height).
# This follows conventions in visualization, which unfortunately are different from those of linear algebra.
plt.rcParams.update({"figure.autolayout":True})

fig,ax=plt.subplots(figsize=(8,4))
ax.barh(group_names,group_data)
labels=ax.get_xticklabels()
plt.setp(labels,rotation=45,horizontalalignment="right")
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
plt.show()
```

```python
def currency(x, pos):
    """The two args are the value and tick position"""
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s

formatter = FuncFormatter(currency)

fig, ax = plt.subplots(figsize=(6,6))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')

ax.xaxis.set_major_formatter(formatter)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# Add a vertical line, here we set the style in the function call
ax.axvline(group_mean, ls='--', color='r')

# Annotate new companies
for group in [3, 5, 8]:
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")
    # here group is the number corresponding to y axis (top to bottom).
    # 145000 is the value of x axis.

# Now we'll move our title up since it's getting a little cramped
ax.title.set(y=1.05)
# changing the y axis of title.

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(formatter)
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
fig.subplots_adjust(right=.1)

plt.show()
```
