```python
# module 16

%matplotlib inline
from cycler import cycler # note this line
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)# endpoint arguement just making the end value exclusive.
yy = np.transpose([np.sin(x + phi) for phi in offsets])
print(yy.shape)
```

```python
# creating cycle of properties.
default_cycler = (cycler(color=['r', 'g', 'b', 'y']) + cycler(linestyle=['-', '--', ':', '-.']))

# changing the properties within rcParams.
plt.rc("lines",linewidth=4)
plt.rc("axes",prop_cycle=default_cycler)
```

```python
custom_cycler=(cycler(color=["c","m","y","k"])+ cycler(linewidth=[2,1,4,3]))
fig,(ax1,ax2)=plt.subplots(nrows=2)
ax1.plot(yy)
ax1.set_title("R B G Y")

ax2.set_prop_cycle(custom_cycler)
ax2.plot(yy)
ax2.set_title("C M Y K")

# Add a bit more space between the two plots.
fig.subplots_adjust(hspace=0.5)
plt.show()
```

```python
# combination of different cycles.

mul_cycle=(cycler(color=["c","m","y","k"])+ cycler(linewidth=[2,1,4,3]))
for cyc in mul_cycle:
    print(cyc)
```

```python
mul2_cycle=(cycler(color=["c","m","y","k"])* cycler(linewidth=[2,1,4,3]))
for cyc in mul2_cycle:
    print(cyc)
```
