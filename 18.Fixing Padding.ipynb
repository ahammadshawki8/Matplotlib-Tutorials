```python
import matplotlib.pyplot as plt
import numpy as np


def example_plot(ax, fontsize=12):
    ax.plot([1, 2])

    ax.locator_params(nbins=3)
    ax.set_xlabel('x-label', fontsize=fontsize)
    ax.set_ylabel('y-label', fontsize=fontsize)
    ax.set_title('Title', fontsize=fontsize)

plt.close('all')
fig, ax = plt.subplots()
example_plot(ax, fontsize=25)
```

```python
fig,ax1= plt.subplots()
example_plot(ax1, fontsize=18)
plt.tight_layout()
```

```python
fig2,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)
```

```python
fig2,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)
plt.tight_layout()
```

```python
# tight_layout() can take keyword arguments of pad, w_pad and h_pad.
# These control the extra padding around the figure border and between subplots.
# The pads are specified in "fraction of fontsize"
fig2,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)
plt.tight_layout(pad=0.4, w_pad=1, h_pad=2)
# pad tighten the graph from both side.
# w_pad tighten the graph from vertical side.
# h_pad tighten the graph from horizontal side.
```

```python
arr = np.arange(64).reshape((8, 8))

plt.close('all')
fig = plt.figure(figsize=(5, 4))

ax = plt.subplot(111)
im = ax.imshow(arr, interpolation="none")

plt.tight_layout()
```
